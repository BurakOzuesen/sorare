```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("soraredata (5).csv")
myData = pd.DataFrame(columns=["player_name", "position", "age", "score", "price"])
```

```python
myData
```

```python
pos_age = data.iloc[:,2]
```

```python
for index, element in data.iterrows():
    player_name = element.values[1]
    position = element.values[2].replace(" ", "").split("-")[0]
    age = element.values[2].replace(" ", "").split("-")[1]
    score = float(element.values[3])
    price = element.values[4]
    if price == 'No sale':
        continue
    else:
        price = float(str(price[2:]).replace(",", "."))

    new_row = {'player_name': player_name, 'position': position, 'age': age, 'score': score, 'price': price, 'diff': 0}

    # Yeni satırı DataFrame'e ekleyin
    myData = myData.append(new_row, ignore_index=True)
```

```python
myData
```

```python
fw_dataset = myData.loc[myData['position'] == "Forward"].reset_index()
mid_dataset = myData.loc[myData['position'] == "Midfielder"].reset_index()
def_dataset = myData.loc[myData['position'] == "Defender"].reset_index()
gk_dataset = myData.loc[myData['position'] == "Goalkeeper"].reset_index()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

scalerX = StandardScaler()
scalerY = StandardScaler()
model = RandomForestRegressor()


X = myData[['age', 'score']]
y = myData[['price']]
X = scalerX.fit_transform(X)
y = scalerY.fit_transform(y)
```

```python
model.fit(X, y)
```

```python
preds = model.predict(X)
y_new_inverse = scalerY.inverse_transform(preds)
y_new_inverse
```

```python
y_new_inverse
```

```python
diff = y_new_inverse - y # çok olan iyidir
type(diff)
myData['diff'] = diff
```

```python
valid_count = 0

team_comp = pd.DataFrame(columns=["p1", "p2", "p3", "p4", "score", "age", "price"])

while True:
    random_data = myData.sample(n=4)
    
    cost = random_data["price"].sum()
    score = random_data["score"].sum()
    age = random_data["age"].sum()
    name_arr = list(random_data["player_name"])

    pos_array = list(random_data["position"])

    gk_count = pos_array.count("Goalkeeper")
    def_count = pos_array.count("Defender")
    mf_count = pos_array.count("Midfielder")
    fw_count = pos_array.count("Forward")

    if def_count == 0 or mf_count > 1 or fw_count == 0:
        continue

    if gk_count != 1:
        continue
    
    if score > 195:
        continue
    
    if cost > 0.0404:
        continue
    

    new_row = {'p1': name_arr[0], 'p2': name_arr[1], 'p3': name_arr[2], 'p4': name_arr[3], 'score': score, 'age': age, "price": cost, "diff": 0}

    team_comp = team_comp.append(new_row, ignore_index=True)

    print(random_data)
    print("Score", score)
    print("Cost", cost)
    print()

    valid_count += 1
    if valid_count > 99:
        break
```

```python
team_comp
```

```python
model = RandomForestRegressor()

X = team_comp[['age', 'score']]
y = team_comp[['price']]

model.fit(X, y)

preds = model.predict(X).reshape(100,1)

diff = preds - y # çok olan iyidir
team_comp['diff'] = diff
```
